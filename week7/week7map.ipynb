```python
hiroshima_pop = pd.read_csv('hiroshima.csv')
```

```python
hiroshima_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
hiroshima_ku
```

```python
hiroshima_pop.info()
```

```python
hiroshima_pop['KEY_CODE'] = hiroshima_pop['KEY_CODE'].astype(int)
```

```python
hiroshima_ku['KEY_CODE'] = hiroshima_ku['KEY_CODE'].astype(int)
```

```python
hiroshima_ku.info()
```

```python
hiroshima_ku_pop = hiroshima_ku.merge(hiroshima_pop,on='KEY_CODE')
```

```python
hiroshima_ku_pop
```

```python
hiroshima_ku_pop.info()
```

```python
hiroshima_ku_pop['人口総数'] = hiroshima_ku_pop['人口総数'].astype(int)
```

```python
hiroshima_ku_pop.info()
```

```python
hiroshima_ku_pop.plot(column='人口総数',legend=True)
```

```python
hiroshima_ku_pop.plot(figsize=(10,10),
                  column='人口総数',
                  legend=True)
```

```python
hiroshima_ku_pop.plot(figsize=(10,10),
                  column='人口総数',
                  legend=True,
                  cmap='seismic')
```
